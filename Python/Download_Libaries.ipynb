```python
list_of_download = [
    'https://repo1.maven.org/maven2/com/google/code/gson/gson/',
    'https://repo1.maven.org/maven2/org/mockito/mockito-all/',
    'https://repo1.maven.org/maven2/org/jsoup/jsoup/',
    'https://repo1.maven.org/maven2/com/fasterxml/jackson/core/jackson-core/',
    'https://repo1.maven.org/maven2/log4j/log4j/',
    'https://repo1.maven.org/maven2/org/slf4j/slf4j-api/',
    'https://repo1.maven.org/maven2/org/apache/commons/commons-lang3/',
    'https://repo1.maven.org/maven2/com/google/guava/guava/',
    'https://repo1.maven.org/maven2/org/apache/httpcomponents/httpclient/',
    'https://repo1.maven.org/maven2/xerces/xercesImpl/',
    'https://repo1.maven.org/maven2/javax/xml/bind/jaxb-api/',
    'https://repo1.maven.org/maven2/org/apache/poi/poi/',
    'https://repo1.maven.org/maven2/org/javassist/javassist/',
    'https://repo1.maven.org/maven2/cglib/cglib/',
    'https://repo1.maven.org/maven2/commons-dbcp/commons-dbcp/'
]
```

```python
import json
import requests
from datetime import datetime, timezone

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/99.0.4844.84 Safari/537.36',
}
i = 0
library_list = []
for li in list_of_categories:
  page = requests.get(li, headers=headers)
  x = page.content
  y = json.loads(x)
  for x in y["response"]["docs"]:
    dt_object = datetime.fromtimestamp(int(x["timestamp"])/1000)
    library_list.append({"link" : list_of_download[i] + x["v"], "year" : dt_object.year , "name" : x["a"], "version": x["v"]})
  i = i+1
print(library_list)
```

```python
for link in library_list:
  links = requests.get(link["link"], headers=headers)
  # print(link["link"])
  from lxml import html
  files_tree = html.fromstring(links.content)
  all_links = files_tree.xpath('/html/body/main/pre/a')
  for l in all_links:
    if l.attrib['href'].endswith("sources.jar"):
      # print(l.attrib['href'])
      jar_file = requests.get(link["link"] + "/" + l.attrib['href'])
      # print(link["link"] + "/" + l.attrib['href'])
      file_name = str(link["name"])+ "_" + str(link["year"]) + "_" + str(link["version"]) + ".jar"
      #print(file_name)
      with open(file_name ,'wb') as f:
        f.write(jar_file.content)
    
  #print(links)
```
